**Principal Component Analysis with Python and numpy**



```python
import numpy as np
from numpy import linalg
```

```python
#Caricare il dataset
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
print('Dataset: ', iris.data)
print('target: ', iris.target)
```

```python
#Creare una variabile che contiene i nostri dati
X = iris.data
```

```python
#Centrare i dati
X_mean = np.mean(X)
X_centered = X - X_mean
```

```python
#Matrice di covarianza
cov_matrix = np.cov(X_centered, rowvar=False)
```

```python
#Auovalori e autovettori
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
```

```python
print('autovalori: ', eigenvalues)
print('autovettori: ', eigenvectors)
```

```python
#Ordinare gli indici degli autovalori in ordine decrescente
sorted_eigenvalues_indexes = np.argsort(eigenvalues)[::-1]
```

```python
#Ordinare gli autovettori in base agli indici degli autovalori corrispondenti
sorted_eigenvectors = eigenvectors[:, sorted_eigenvalues_indexes]
```

```python
#Selezionare gli autovettori
selected_eigenvectors = sorted_eigenvectors[:, :2]
```

```python
#Calcolare le componenti principali
pc = np.dot(X_centered, selected_eigenvectors)
```

```python
#proiettare i dati nel sottospazio delle componenti principali
projected_data = np.dot(pc, selected_eigenvectors.T)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_centered[:,0], X_centered[:,1])
plt.xlabel('lunghezza del petalo')
plt.ylabel('larghezza del petalo')
```

```python
plt.figure() #serve a separare le due immagini
plt.scatter(projected_data[:,0], projected_data[:,1])
plt.xlabel('lunghezza del petalo')
plt.ylabel('larghezza del petalo')
```

```python
total_variance = np.sum(eigenvalues)
sum_selected_eigenvalues = np.sum(eigenvalues[:2])
```

```python
#Varianza spiegata
explained_variance = sum_selected_eigenvalues/total_variance
print('varianza spiegata: ', explained_variance)
```
